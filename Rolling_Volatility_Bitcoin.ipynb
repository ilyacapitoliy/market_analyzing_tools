### Bitcoin Rolling Volatility

###### The code helps to know historical volatility of Bitcoin


```python
import matplotlib.pyplot as plt
import yfinance as yf

#TODAY
from datetime import date
today = str(date.today())

# Fetch Bitcoin data from Yahoo Finance
btc = yf.download('BTC-USD', start='2014-01-01', end=today)

# Calculate daily returns
btc['Returns'] = btc['Close'].pct_change() * 100

# Define rolling window size
window_size = 90  # You can adjust this according to your preference

# Calculate rolling volatility
btc['Rolling Volatility'] = btc['Returns'].rolling(window=window_size).std()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(btc.index, btc['Rolling Volatility'], color='blue', label=f'Rolling Volatility ({window_size} days)')
plt.title('Bitcoin Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```
